```python
while True:
  n = int(input("What's n: "))
  if n>0:
    break

for _ in range(n):
  print("Meow")
```

```python
def main():
  number = get_number()
  meow(number)

def get_number():
  while True:
    n = int(input("What's n: "))
    if n>0:
      break
  return n

def meow(n):
  for _ in range(n):
    print("meow")

main()
```

```python
def main():
  print_column(3)
  print_row(3)
  print_square(3)

def print_column(height):
  for _ in range(height):
    print("#")

def print_row(width):
    print("$" * width)

def print_square(size):
  for i in range(size):
    for j in range(size):
      print("*", end="")
    print()

main()
```
