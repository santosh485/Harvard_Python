```python

x = float(input("What's x: "))
y = float(input("What's y: "))

#rounding
z = round(x+y)
#formating with coma
print(f"{z:,}")

#rounding in 3 digit
z = round(x/y, 3)
print(z)
#or
print(f"{z:.2f}")
```
